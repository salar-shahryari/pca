```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
y
```

```python
# Standardize the data
X_std = StandardScaler().fit_transform(X)

# Apply PCA (reduce to 2 dimensions)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_std)

# Create a DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components,
                            columns=['Principal Component 1', 'Principal Component 2'])
principal_df['target'] = y

# Plot the principal components
plt.figure(figsize=(8, 8))
colors = ['r', 'g', 'b']
for target, color in zip([0, 1, 2], colors):
    indices_to_keep = principal_df['target'] == target
    plt.scatter(principal_df.loc[indices_to_keep, 'Principal Component 1'],
                principal_df.loc[indices_to_keep, 'Principal Component 2'],
                c=color,
                s=50)
plt.xlabel('Principal Component 1', fontsize=15)
plt.ylabel('Principal Component 2', fontsize=15)
plt.title('2 component PCA', fontsize=20)
plt.legend(target_names)
plt.grid()
plt.show()
```
